Transformando a coluna 'overall_rating' de 1 a 5 para entradas de 0 para reviews de 1 a 3 e 1 para reviews de 4 e 5 


```python
sentiments = []
for rating in df['overall_rating']:

  if rating > 3:
    sentiments.append(1)

  else: 
    sentiments.append(0)


df['sentiment'] = sentiments
```

```python
plt.figure(figsize=(10,8))
sns.set_style('whitegrid')
sns.set_context('poster')
sns.countplot(df['sentiment'])
plt.show()
```
