```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
from PIL import Image
from wordcloud import WordCloud

text = " ".join(s.lower() for s in df['review_comentarios'])
wordcloud = WordCloud(background_color="orange",
                      width=1600, height=800).generate(text)
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python

stopwords_pt = stopwords.words("portuguese")
stopwords_pt.remove('não')
stopwords_pt.remove('nem')


text = " ".join(s.lower() for s in df['review_comentarios'])
wordcloud = WordCloud(stopwords=stopwords_pt, background_color="purple",
                      width=1600, height=800).generate(text)
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

```
