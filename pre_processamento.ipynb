```python
import re
import swifter 
import string
```

```python
def regex_texto(df):
    rv = df.apply(lambda x: re.sub("http\S+", '', str(x)))
    rv = rv.swifter.apply(lambda x: re.sub(u'[^a-zA-Z0-9áéíóúÁÉÍÓÚâêîôÂÊÎÔãõÃÕçÇ: ]', '',x))
    rv = rv.swifter.apply(lambda x: re.sub(' +', ' ', x))
    rv = rv.swifter.apply(lambda x: re.sub('(#[A-Za-z]+[A-Za-z0-9-_]+)', '', x)) 
    rv = rv.swifter.apply(lambda x: re.sub('(@[A-Za-z]+[A-za-z0-9-_]+)', '', x)) 
    rv = rv.swifter.apply(lambda x: re.sub('rt', '', x)) 
    #rv = rv.swifter.apply(lambda x: re.sub('[A-Z]', '',x))
    rv = rv.swifter.apply(lambda x: re.sub(r'[0-9]+', '', x)) 
    rv = rv.swifter.apply(lambda x: ''.join([i for i in x if i not in string.punctuation]))
    return rv
```

```python
df['reviews_limpos'] = regex_texto(df['review_text'])
df.head()
```

```python
df['review_title_limpo'] = regex_texto(df['review_title'])
df.head()
```

```python
df.drop(['site_category_lv1', 'site_category_lv2', 'recommend_to_a_friend', 'product_brand', 'submission_date', 'product_id'],1, inplace = True)
```

```python
df.drop(['reviewer_id'],1, inplace = True)
```

```python
df.drop(['product_name', 'reviewer_birth_year'],1, inplace = True)
```

```python
df.drop(['reviewer_gender','reviewer_state'],1, inplace = True)
```

```python
df['review_comentarios'] = df['review_title_limpo'] + ' ' + df['reviews_limpos']
df.head()
```
